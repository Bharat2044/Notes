## Understanding Distributed Computing

### What is Distributed Computing?

Distributed computing involves breaking down computational tasks across multiple computers, enhancing performance and scalability. This is particularly crucial in the realm of machine learning where large datasets and complex models demand significant computational resources.

### Key Concepts: Node, Communication and Concurrency

1. **Node:** A node represents an individual computer or server within a distributed system. In machine learning systems, nodes could be servers with powerful GPUs, allowing for parallel processing of data.

    Example: In Google's TensorFlow Extended (TFX), nodes can be individual servers responsible for distributed training of machine learning models.

````python
file = '../img/tfx.png'
display(Image(filename=file, embed=True, width=1000, height=800))
````

2. **Communication:** Nodes in a distributed system communicate to share information, ensuring that tasks are coordinated and results are aggregated. Communication can occur through direct messaging or shared data storage.

    Example: Apache Kafka is often used in distributed machine learning systems for real-time data streaming and communication between various components.

````python
file='../img/kafka.png'
display(Image(filename=file, embed=True, width=600, height=500))
````

3. **Concurrency:** Concurrency involves the simultaneous execution of multiple tasks. In machine learning, this is vital for processing large datasets and training models efficiently.

    Example: Apache Spark is used for distributed data processing in systems like Databricks, enabling concurrent processing of large-scale machine learning tasks.

````python
file='../img/spark.png'
display(Image(filename=file, embed=True, width=800, height=500))
````

### Why Distributed Computing?

- **Scalability:** Machine learning models often require vast amounts of data for training. Distributed computing allows for easy scalability by adding more nodes to the system.

    Example: Facebook's Prophet, a time series forecasting tool, utilizes distributed computing for scalability, enabling it to handle large-scale time series data.
    

- **Fault Tolerance:** In distributed systems, components can fail, but the system remains operational. This is critical for ensuring uninterrupted machine learning tasks, even in the face of hardware failures.

    Example: Kubernetes, used by many companies, provides fault tolerance by automatically rescheduling tasks on healthy nodes when a node fails.

- **Performance:** Distributed computing enhances performance by parallelizing tasks, reducing processing time for machine learning workflows.

    Example: Apache Spark MLLib employs distributed computing for model training, allowing users to train large-scale machine learning models efficiently.

## Docker and Microservices

### Microservices Architecture:

Microservices architecture is an approach to designing and building software applications as a collection of small, independent services. Each service is a standalone unit, and they communicate with each other through well-defined APIs. This architecture provides flexibility, scalability, and ease of maintenance.

- **Key Characteristics**:
    - Modularity: Services are independent, allowing changes to one service without affecting others.
    - Scalability: Individual services can be scaled independently based on demand.
    - Resilience: Failure in one service doesn't necessarily impact the entire system.
    - Technology Diversity: Different services can use different technologies and programming languages.

### Docker in Microservices:

Docker is a containerization platform that allows developers to package applications and their dependencies into lightweight containers. Each container runs in its own isolated environment, ensuring consistency across different stages of development and deployment.

#### How Docker Facilitates Microservices:

- Isolation: Each microservice runs in its own Docker container, preventing interference with other services.

- Portability: Containers encapsulate everything needed to run a service, making it easy to move across different environments.

- Consistency: Docker ensures a consistent environment for microservices, reducing the "it works on my machine" problem.

#### Real-world Examples in Machine Learning Systems:

1. **Netflix - Mantis**:
Netflix employs a microservices architecture for Mantis, its real-time stream processing system. Each microservice, responsible for specific tasks in stream processing, runs in its own Docker container. This approach allows Netflix to scale components independently, enhancing the overall efficiency of real-time data processing.

2. **Uber - Michelangelo**:
Uber's machine learning platform, Michelangelo, adopts microservices for various components such as model training, serving, and monitoring. Docker containers enable seamless deployment and scaling of these microservices. For instance, the serving component runs as an independent microservice, allowing it to scale horizontally to handle varying inference loads.

3. **Airbnb - Bighead**:
Airbnb's Bighead is a machine learning platform with a microservices architecture. Docker containers are used to encapsulate individual components, such as feature engineering, model training, and deployment. This approach enables Airbnb's data science teams to work on and deploy their models independently.

#### Advantages of Docker in Machine Learning Microservices:

- Dependency Management: Docker ensures that each microservice has its dependencies encapsulated, reducing conflicts and ensuring reproducibility.

- Efficient Resource Utilization: Containers share the host OS kernel, leading to efficient resource utilization and faster startup times.

- Scalability: Docker containers can be easily replicated, making it simple to scale individual microservices horizontally as demand increases.

Kubernetes (K8s) is an open-source container orchestration platform that automates the deployment, scaling, and management of containerized applications. It abstracts away the underlying infrastructure, providing a consistent and scalable environment for deploying and running applications.

#### 1. Pods

- **Definition:** The smallest deployable units in Kubernetes. Pods can contain one or more containers sharing the same network namespace.

- **Example:**
  ```yaml
  apiVersion: v1
  kind: Pod
  metadata:
    name: ml-app
  spec:
    containers:
    - name: python-app
      image: python-app:latest
    - name: redis-db
      image: redis:latest
  ```

#### 2. Nodes

- **Definition:** Physical or virtual machines in a Kubernetes cluster where pods are deployed. Nodes can be virtual machines in the cloud or physical machines in an on-premises data center.

- **Example:** In a Kubernetes cluster, you may have multiple nodes, each representing a server. Nodes collaborate to run and manage pods.

#### 3. Cluster

- **Definition:** A collection of nodes that work together to run containerized applications. A cluster consists of a control plane and multiple nodes.

- **Example:** A Kubernetes cluster might have three nodes for running pods. The control plane manages the overall state of the cluster, ensuring that the desired state matches the actual state.

#### 4. Control Plane

- **Definition:** The set of components that manage the overall state of the Kubernetes cluster. It includes the API server, controller manager, etcd, and scheduler.

- **Example:** The control plane ensures that the declared configuration (defined in YAML manifests) is maintained. For example, if a pod fails, the control plane will attempt to reschedule it on a healthy node.

#### 5. Deployment

- **Definition:** A higher-level abstraction that allows you to declare and manage a desired state for your application. Deployments ensure that the specified number of replicas of a pod are running.

- **Example:**
  ```yaml
  apiVersion: apps/v1
  kind: Deployment
  metadata:
    name: ml-service
  spec:
    replicas: 3
    template:
      metadata:
        labels:
          app: ml-service
      spec:
        containers:
        - name: ml-container
          image: ml-service:latest
  ```

#### 6. Service [ Not covering ]

- **Definition:** An abstraction that defines a logical set of pods and a policy to access them. Services provide stable endpoints for communication between pods.

- **Example:**
  ```yaml
  apiVersion: v1
  kind: Service
  metadata:
    name: ml-service
  spec:
    selector:
      app: ml-service
    ports:
      - protocol: TCP
        port: 80
        targetPort: 5000
  ```

#### 7. ConfigMap [ Not covering ]

- **Definition:** An API object used to decouple configuration data from container images. ConfigMaps allow you to separate configuration details from application code.

- **Example:**
  ```yaml
  apiVersion: v1
  kind: ConfigMap
  metadata:
    name: ml-config
  data:
    hyperparam1: "value1"
    hyperparam2: "value2"
  ```

These key concepts form the foundation of Kubernetes, enabling the efficient deployment, scaling, and management of containerized applications. In the context of machine learning, Kubernetes provides a robust infrastructure for deploying and orchestrating ML models and services at scale.

## Deploying Applications with Kubernetes

#### YAML Manifests:

Kubernetes utilizes YAML files, referred to as manifests, to articulate the configuration of resources within the cluster. These manifests provide a declarative way to express the desired state of the application.

**Example:**
Consider a YAML manifest for deploying a simple machine learning service as a Kubernetes Deployment:

```yaml
apiVersion: apps/v1
kind: Deployment
metadata:
  name: ml-service
spec:
  replicas: 3
  template:
    metadata:
      labels:
        app: ml-service
    spec:
      containers:
      - name: ml-container
        image: your-docker-registry/ml-service:latest
      resources:
          limits:
            memory: "1024Mi"
            cpu: "2000m"
        ports:
        - containerPort: 8080
```

In this example, a Kubernetes Deployment named `ml-service` is defined with three replicas. The template specifies the pod's specifications, including the container running the machine learning service.

#### Deploying with `kubectl`:

Once the YAML manifest is created, `kubectl` is the command-line tool used to interact with the Kubernetes cluster. It can be employed to apply the configuration defined in the manifest and manage the deployed resources.

1. **Applying a Deployment:**
   ```bash
   kubectl apply -f deployment.yaml
   ```
   This command creates or updates the resources defined in the YAML manifest, specifically the Deployment for the machine learning service.

2. **Scaling:**
   ```bash
   kubectl scale --replicas=5 deployment/ml-service
   ```
   This scales the number of replicas for the `ml-service` Deployment to 5, allowing for horizontal scaling of the machine learning service.

3. **Checking Deployment Status:**
   ```bash
   kubectl get deployments
   ```
   This command provides information about the status of Deployments, including the number of replicas, available replicas, and current state.
